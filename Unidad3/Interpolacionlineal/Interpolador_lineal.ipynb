```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos=np.loadtxt("datos.dat")
x_list=[]
y_list=[]
for i in range(0,datos.shape[0]):
    x_list.append(datos[i][0])
    y_list.append(datos[i][1])
    
def Inlineal(x_list,y_list,x):
    for i in range(0,len(x_list)-1):
        if (x>=x_list[i] and x<=x_list[i+1]):
            value=(y_list[i+1]-y_list[i])/(x_list[i+1]-x_list[i])*(x-x_list[i])+y_list[i]
            return value
x_interpol=np.linspace(min(x_list),max(x_list),100)
y_interpol=[]
for i in range(0,len(x_interpol)):
    y_interpol.append(Inlineal(x_list,y_list,x_interpol[i]))
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x_list,y_list,'o',markersize=10,label='Datos discretos')
ax.plot(x_interpol,y_interpol,'>',color='purple',markersize=5,label='Puntos interpolados')
ax.plot(x_interpol,np.sin(x_interpol),color='green',label='$sin(x)$')
ax.legend()
ax.grid()
plt.show()
```
